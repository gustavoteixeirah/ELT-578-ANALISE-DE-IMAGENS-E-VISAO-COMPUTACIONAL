```python
import cv2

#Coordenada [x,y] quando acionado o botão direito
XY = list()

#função armazenará uma coordenada toda vez que o botão direito for acionado,
# até uma tecla qualquer for acionada
def direito_down(evento, x, y, flags, params):

    # evento igual a 2 quando acionado botão direito
    if evento == 2:
        global XY
        XY.append([x, y])


caminho_imagem = '.\ImagensB\Claro2247.png'
I = cv2.imread(caminho_imagem,0)


# ajustar uma imagem para uma janela onde a maior dimensão terá
# 512 pixels na tela
escala = min(512./I.shape[0], 512./I.shape[1])
largura_window = int(I.shape[1] * escala)
altura_window = int(I.shape[0] * escala)
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.resizeWindow('image', largura_window, altura_window)

# Chamada toda vez que uma tecla do mouse é acionada
cv2.setMouseCallback('image', direito_down)

cv2.imshow('image', I)
cv2.waitKey(0)
cv2.destroyAllWindows()

print (XY)
```
