```python
#Importando bibliotecas
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

#Imagem colorida
img = cv2.imread('/content/drive/MyDrive/ELT578/SEMANA_1/flower.jpeg')

cv2_imshow(img)

print ('Dimensoes da Imagem Flor (linhas, colunas e bandas) = ', img.shape)
print ('Tamanho (px) da Imagem Flor = ', img.size)
print ('Valores das bandas BGR na linha 100 e coluna 100 da Flor = ', img[99,99])
print ('Valor da banda azul na linha 100 e coluna 100 da Flor = ', img[99,99,0])
```

```python
img2 = cv2.imread('/content/drive/MyDrive/ELT578/SEMANA_1/flower.jpeg', cv2.IMREAD_GRAYSCALE)
cv2_imshow(img2)
print ('Dimensoes da Imagem Flor (linhas, colunas e bandas) = ', img2.shape)
print ('Tamanho (px) da Imagem Flor = ', img2.size)
print ('Valores do px em tons de cinza linha 100 e coluna 100 da Flor = ', img2[99,99])
```

```python
#Visualizando imagens grandes
from matplotlib import pyplot as plt

plt.imshow(img)
plt.xticks([]), plt.yticks([]) # esconder os valors de Ox e Oy #esconder o eixo
plt.show()

#O OpenCV lê a imagem no sistema BGR, o Matplotlib trabalha no sistema RGB. Então, temos que converter a imagem no momento de visualizar dentro do Matplolib.
```

```python
#Visualizando imagens grandes
from matplotlib import pyplot as plt
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Visualizar as duas imagens ao mesmo tempo
titulos = ['Flor RGB', 'Flor Gray']
imagens = [img, img2]

#0,1 em uma lista
for i in list(range(2)): # i varia de 0 a 1
    plt.subplot (1,2,i+1), plt.imshow(cv2.cvtColor(imagens[i], cv2.COLOR_BGR2RGB))
    plt.title(titulos[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Salvando Flor Gray
cv2.imwrite('/content/drive/MyDrive/ELT578/SEMANA_1/flowerGray.jpeg', cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
```

```python
#Qual é o tamanho de armazenamento teórico da imagem?

# Inicialmente, vamos determinar a profundidade do pixel da imagem
print(img.dtype)
print(img.shape)
#Considerando que a imagem tem 2048 linhas e 946 colunas, temos:
npx=2048*946

#Pela profundidade do pixel, temos que são necessários 8 bits para armazenar esse valor ou 1 byte,
#logo o número de bytes na imagem coincide com número de pixels, pois a imagem possui apenas 1 banda já que não é colorida.
#Sabendo que 1kB = 1024 bytes, temos:


#Considerando que cada byte de armazenamento é equivalente a 8 bits e que a imagem "Flor" possuí 3 bandas, ou seja, 24bits.
#O tamanho de armazenamento digital em kilobytes (1024bytes) da imagem é dado por:
size=npx/(1024*3)
print('O tamanho teórico da imagem é %1.3f kB' %size)

import os
print('O tamanho real da imagem é de %1.3f kB' % (os.path.getsize('/content/drive/MyDrive/ELT578/SEMANA_1/flower.jpeg')/1024))
```

```python
#Salvando a imagem em formato JPEG com 3 diferentes níveis de qualidade
cv2.imwrite('/content/drive/MyDrive/ELT578/SEMANA_1/flower25.jpeg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 25])
cv2.imwrite('/content/drive/MyDrive/ELT578/SEMANA_1/flower50.jpeg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 50])
cv2.imwrite('/content/drive/MyDrive/ELT578/SEMANA_1/flower75.jpeg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 75])
cv2.imwrite('/content/drive/MyDrive/ELT578/SEMANA_1/flower100.jpeg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 100])


#Visualizar as 4 imagens ao mesmo tempo
img25 = cv2.imread('/content/drive/MyDrive/ELT578/SEMANA_1/flower25.jpeg')
img50 = cv2.imread('/content/drive/MyDrive/ELT578/SEMANA_1/flower50.jpeg')
img75 = cv2.imread('/content/drive/MyDrive/ELT578/SEMANA_1/flower75.jpeg')
img100 = cv2.imread('/content/drive/MyDrive/ELT578/SEMANA_1/flower100.jpeg')

title = ['Flor 25', 'Flor 50', 'Flor 75', 'Flor 100']
imagens = [img25, img50, img75, img100]

for i in list(range(4)):
    plt.subplot (2,2,i+1), plt.imshow(cv2.cvtColor(imagens[i], cv2.COLOR_BGR2RGB))
    plt.title(title[i])
    plt.xticks([]), plt.yticks([])
plt.show()

print(img25.shape)
print(img50.shape)
print(img75.shape)
print(img100.shape)

#As imagens foram salvas em quatro padrões de qualidade diferentes (25%, 50%, 75% e 100%), podemos observar que todas as imagens possuem a mesma resolução da imagem, bem como a mesma resolução de brilho (24 bits, dado que uma imagem RGB possuí 3 bandas).
#A mudança quantitativa é então observada a nível de tamanho de armazenamento (kBytes).
#Os formatos de imagem são padrões de codificação, compressão e definição que determinam o tipo de arquivo gráfico, como JPEG, PNG, BMP, GIF e TIFF.
#Muitas vezes, o formato escolhido é uma questão de equilíbrio entre a qualidade e o tamanho da imagem, de acordo com a aplicação desejada.
```
